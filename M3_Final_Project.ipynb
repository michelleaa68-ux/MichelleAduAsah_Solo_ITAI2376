```python
# ============================================================
# MIRA — Multi-Agent Event Decor Intelligence System
# Michelle Adu-Asah | ITAI 2376
#
# MIRA is a three-agent AI system built for luxury event
# decor companies and their clients.
#
# MIRA  — The client-facing chatbot. Handles all general
#          and technical questions such as pricing, location,
#          appointments, delivery, and event details.
#
# MIZEL — The creative director. Clients describe their
#          vision in their own words and MIZEL responds with
#          a full color scheme and mood board tailored to
#          their event.
#
# MILO  — Internal warehouse tool for decor workers only.
#          Protected by a passcode that clients cannot access.
#          Workers use it to check inventory availability.
# ============================================================

def mira(question):
    question = question.lower()

    if "price" in question or "cost" in question or "how much" in question:
        answer = (
            "Our packages start at $2,500 for intimate gatherings. "
            "The Elegance package is $2,500–$5,000, the Luxe package is $5,000–$12,000, "
            "and the Couture package is fully custom quoted based on your vision and guest count."
        )

    elif "chandelier" in question or "how tall" in question or "height" in question:
        answer = (
            "Our chandeliers range from 24 inches to 48 inches in diameter and hang at a "
            "standard drop of 6 to 10 feet from the ceiling. We adjust the height based on "
            "your venue's ceiling clearance during the site visit."
        )

    elif "extra table" in question or "table setting" in question or "charger" in question or "linen" in question:
        answer = (
            "Additional table settings such as charger plates, linen napkins, and centerpiece upgrades "
            "are available as add-ons. Charger plates start at $2 each, linen upgrades start at $15 per table, "
            "and centerpiece enhancements are quoted based on your floral selections."
        )

    elif "appointment" in question or "meet" in question or "consultation" in question or "schedule" in question:
        answer = (
            "We would love to meet with you! You can book a free 30-minute discovery call "
            "by contacting us at hello@mira-events.com or calling (713) 555-0192. "
            "We recommend scheduling at least 8 weeks before your event date to secure availability."
        )

    elif "location" in question or "where are you" in question or "address" in question:
        answer = (
            "We are based in Houston, TX and service the greater Houston metro area including "
            "Sugar Land, Katy, Pearland, and The Woodlands. We also travel for destination events — "
            "just ask about our travel rates!"
        )

    elif "transport" in question or "delivery" in question or "setup" in question or "install" in question:
        answer = (
            "Our team handles full delivery, setup, and breakdown of all decor items. "
            "Transport and setup fees start at $350 and vary based on venue distance and event size. "
            "Everything is included in your final quote — no surprise fees on event day."
        )

    elif "book" in question or "reserve" in question or "deposit" in question:
        answer = (
            "To reserve your date, we require a signed contract and a 30% deposit. "
            "The remaining balance is due two weeks before your event. "
            "Reach out to us to get started — popular dates book quickly!"
        )

    elif "cancel" in question or "refund" in question:
        answer = (
            "Cancellations made 60 or more days before the event receive a full refund minus a $200 processing fee. "
            "Cancellations within 30 days are non-refundable, but store credit may be issued at our discretion. "
            "Please contact us as early as possible if your plans change."
        )

    else:
        answer = (
            "Thank you so much for reaching out! Our team is available Monday through Saturday, 9AM to 6PM CST. "
            "For after-hours questions, email us at hello@mira-events.com and we will get back to you within 24 hours."
        )

    print("=" * 55)
    print("  MIRA — Client Assistant")
    print("=" * 55)
    print(f"  Client asked: {question}")
    print()
    print(f"  MIRA says:")
    print(f"  {answer}")
    print()


# ---- AGENT 2: MIZEL ----
# The client describes their vision in their own words.
# MIZEL reads it and suggests a full color scheme and mood board.

def mizel(client_vision):
    vision = client_vision.lower()

    print("=" * 55)
    print("  MIZEL — Creative Visual Agent")
    print("=" * 55)
    print(f"  Client vision: {client_vision}")
    print()

    # MIZEL reads the vision and builds a color scheme + mood board

    if "white" in vision and ("candle" in vision or "garden" in vision or "flower" in vision):
        color_scheme = "Ivory White, Sage Green, and Soft Gold"
        mood_board = (
            "For your guest tables, we recommend ivory linen tablecloths with sage green "
            "napkin folds and soft gold charger plates. Low garden-style floral centerpieces "
            "with white ranunculus, eucalyptus, and trailing greenery will complement your stage perfectly. "
            "Add slim gold pillar candles at varying heights across the tables to mirror the candlelit "
            "atmosphere on stage. The overall feel will be organic, romantic, and cohesive."
        )

    elif "blush" in vision or "pink" in vision or "romantic" in vision or "rose" in vision:
        color_scheme = "Blush Pink, Champagne Gold, and Dusty Rose"
        mood_board = (
            "We suggest blush pink linen with champagne gold charger plates for your guest tables. "
            "Centerpieces of soft pink garden roses, white peonies, and baby's breath in low arrangements "
            "will carry the romantic energy from stage to every corner of the room. "
            "Warm candlelight and champagne ribbon napkin ties will complete the look beautifully."
        )

    elif "royal" in vision or "blue" in vision or "grand" in vision or "regal" in vision:
        color_scheme = "Royal Blue, Gold, and Crisp White"
        mood_board = (
            "For a grand and regal feel, we recommend deep royal blue linen with bright gold charger plates "
            "and white floral centerpieces featuring hydrangeas and white orchids with gold ribbon accents. "
            "Tall centerpiece arrangements will create that majestic ballroom atmosphere. "
            "Gold candelabras at alternating tables will add drama and elegance."
        )

    elif "red" in vision or "burgundy" in vision or "bold" in vision or "luxury" in vision:
        color_scheme = "Deep Burgundy, Forest Green, and Warm Gold"
        mood_board = (
            "Rich burgundy linen with warm gold charger plates will ground your guest tables in luxury. "
            "Centerpieces of deep red roses, burgundy dahlias, and lush greenery in tall gold vases "
            "will create a striking and opulent feel throughout the room. "
            "Dark velvet napkins in burgundy or forest green will add texture and depth to every place setting."
        )

    elif "minimalist" in vision or "simple" in vision or "clean" in vision or "modern" in vision:
        color_scheme = "Crisp White, Slate Grey, and Brushed Silver"
        mood_board = (
            "For a clean and modern look, we suggest white linen with brushed silver charger plates "
            "and minimal low centerpieces featuring white calla lilies and structured greenery. "
            "Simple white taper candles in silver holders will keep the aesthetic sleek and intentional. "
            "Less is more here — every detail should feel purposeful and refined."
        )

    else:
        color_scheme = "Ivory, Soft Gold, and Warm Blush"
        mood_board = (
            "Based on your vision, we suggest a warm and timeless palette of ivory, soft gold, and blush. "
            "Guest tables with ivory linen, gold charger plates, and mixed floral centerpieces in warm tones "
            "will complement almost any stage design beautifully. "
            "We would love to schedule a full consultation to refine this further with fabric swatches and samples!"
        )

    print(f"  Suggested color scheme: {color_scheme}")
    print()
    print(f"  Mood board:")
    print(f"  {mood_board}")
    print()


# ---- AGENT 3: MILO ----
# MILO is for warehouse and decor workers ONLY.
# Workers must enter a passcode to access inventory.

WORKER_PASSCODE = "milo2376"

inventory = {
    "chairs": 150,
    "tables": 20,
    "charger plates": 200,
    "linen tablecloths": 40,
    "linen napkins": 300,
    "silverware sets": 250,
    "floral arches": 3,
    "candelabras": 12,
    "uplighting units": 24,
    "flower wall panels": 8
}

def milo():
    print("=" * 55)
    print("  MILO — Warehouse Inventory System")
    print("  ** FOR DECOR WORKERS ONLY **")
    print("=" * 55)

    code = input("  Enter worker passcode: ")

    if code != WORKER_PASSCODE:
        print()
        print("  Access denied. This system is for authorized workers only.")
        print()
        return

    print()
    print("  Access granted. Current inventory:")
    print()
    print(f"  {'ITEM':<25} {'IN STOCK'}")
    print(f"  {'-'*35}")

    for item, quantity in inventory.items():
        if quantity <= 10:
            flag = "  <-- LOW STOCK"
        else:
            flag = ""
        print(f"  {item.title():<25} {quantity}{flag}")

    print()
    print("  For restocking requests, contact the warehouse manager.")
    print()


# ============================================================
# RUN THE SYSTEM
# ============================================================

print()
print("  Choose who is using the system:")
print("  1 - Client (talk to MIRA)")
print("  2 - Client (describe vision to MIZEL)")
print("  3 - Worker (access MILO inventory)")
print()

choice = input("  Enter 1, 2, or 3: ")

if choice == "1":
    question = input("  What is your question? ")
    print()
    mira(question)

elif choice == "2":
    vision = input("  Describe your event vision: ")
    print()
    mizel(vision)

elif choice == "3":
    print()
    milo()

else:
    print("  Invalid choice. Please run again and enter 1, 2, or 3.")
```
